```python
import pandas as pd
df = pd.read_csv(r'C:\Users\Vishal\Desktop\copper model\Copper_Set.csv')
print(df.head())
```

```python
print(df.isnull().sum())
print(df.describe())
df['material_ref'] = df['material_ref'].replace(r'^00000.*', pd.NA, regex=True)
```

```python
# Fill missing numerical values with the median
df['item_date'].fillna(df['item_date'].median(), inplace=True)
df['customer'].fillna(df['customer'].median(), inplace=True)
df['country'].fillna(df['country'].median(), inplace=True)
df['application'].fillna(df['application'].median(), inplace=True)
df['thickness'].fillna(df['thickness'].median(), inplace=True)
df['selling_price'].fillna(df['selling_price'].median(), inplace=True)

# Fill missing categorical values with the mode
df['status'].fillna(df['status'].mode()[0], inplace=True)
df['material_ref'].fillna(df['material_ref'].mode()[0], inplace=True)

# Drop rows with missing 'id' or 'selling_price' if still any
df.dropna(subset=['id', 'selling_price'], inplace=True)

# Verify if all missing values are handled
print(df.isnull().sum())
```

```python
# Fill the remaining missing value in 'delivery date' with the median
df['delivery date'].fillna(df['delivery date'].median(), inplace=True)

# Verify no missing values
print(df.isnull().sum())
```

```python
# Convert relevant columns to numeric, setting errors='coerce' to convert invalid parsing to NaN
continuous_cols = ['quantity tons', 'thickness', 'width', 'selling_price']

for col in continuous_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Now, let's rerun the outlier capping function
def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Capping the outliers
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    print(f"Outliers capped for {column}")

# Apply to continuous columns again
for col in continuous_cols:
    cap_outliers(df, col)

# Verify the changes
df.describe()
```

```python
import numpy as np  # Import numpy

# Remove negative values in 'quantity tons', replacing them with NaN
df['quantity tons'] = df['quantity tons'].apply(lambda x: x if x >= 0 else np.nan)

# You can do the same for other columns if needed
# For example, if thickness or width can't be negative:
df['thickness'] = df['thickness'].apply(lambda x: x if x >= 0 else np.nan)
df['width'] = df['width'].apply(lambda x: x if x >= 0 else np.nan)

# Now, check if there are any NaN values left in the dataset
print(df.isnull().sum())

# Optionally, you can decide whether to fill or drop rows with NaN
df.dropna(inplace=True)  # This will remove rows with NaN values

# Verify the final dataset
df.describe()
```
